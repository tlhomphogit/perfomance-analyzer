```python
# ---------- ANALYZER MODULE ----------
import os
import csv

class PerfomanceAnalyser():
    def __init__(self):
        self.student_data = []

    def read_from_file(self, filename):
        if not os.path.exists(filename):
            raise FileNotFoundError(f'Error! The file {filename} was not found.')

        with open(filename, mode='r') as file:
            student_results = csv.DictReader(file)
        
            for row in student_results:
                row['score'] = int(row['score'])
                self.student_data.append(row)
                
    def __str__(self):
        return f'{self.student_data}'
    
    def __repr__(self):
        pass

def main():
    filename = 'student_results.csv'
    perfomance = PerfomanceAnalyser()
    perfomance.read_from_file(filename)
    print(perfomance)

if __name__ == '__main__':
    main()
```

```python
# ---------- ANALYZER MODULE ----------
import os
import csv

class PerfomanceAnalyser():
    def __init__(self):
        self.student_data = []

    def read_from_file(self, filename):
        if not os.path.exists(filename):
            raise FileNotFoundError(f'Error! The file {filename} was not found.')

        with open(filename, mode='r') as file:
            student_results = csv.DictReader(file)
        
            for row in student_results:
                row['score'] = int(row['score'])
                self.student_data.append(row)

    
    def calc_avg(self):
        self.scores = [s['score'] for s in self.student_data]
        return round(sum(self.scores) / len(self.scores), 2)
    
    def get_lowest(self):
        return min(self.scores)

    def get_highest(self):
        return max(self.scores)
    
    def get_status(self):
        self.student_data = [
            {'name': s['name'], 
            'surname': s['surname'],
            'score': s['score'], 
            'status': 'pass' if s['score'] >= 50 else 'fail'} for s in self.student_data
        ]


    def get_number_stud(self):
        return len(self.student_data)
        
        
def main():
    filename = 'student_results.csv'
    perfomance = PerfomanceAnalyser()
    perfomance.read_from_file(filename)

    print('*' * 29)
    print('STUDENTS PERFOMANCE ANALYZER')
    print('*' * 29)
    print(f'Avarage Score      : \t{perfomance.calc_avg()}')
    print(f'Highest Score      : \t{perfomance.get_highest()}')
    print(f'Lowest Score       : \t{perfomance.get_lowest()}')
    print(f'Number of Students : \t{perfomance.get_number_stud()}')
    print('*' * 29)

if __name__ == '__main__':
    main()
```

```python
# ---------- CREATE A CSV FILE ----------
import csv

students = [
    {"name": "Tlhompho", "surname": "Ratlhogo", "score": 37},
    {"name": "Moses", "surname": "Mabogwane", "score": 72},
    {"name": "Mmakgomo", "surname": "Seete", "score": 90},
    {"name": "Silas", "surname": "Rakwena", "score": 88},
    {"name": "Kagiso", "surname": "Makgata", "score": 65},
    {"name": "Khothatso", "surname": "Seeme", "score": 45}
]

filename = "student_results.csv"

with open(filename, mode='w', newline='') as file:
    fieldnames = ['name', 'surname', 'score']
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(students)

print(f"File '{filename}' has been created successfully!")
```
