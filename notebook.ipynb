2. ---------- ADD CONTENT TO A LIST OF DICTIONARIES

```python
# ---------- ANALYZER MODULE ----------
import os
import csv

class PerfomanceAnalyser():
    def __init__(self):
        self.student_data = []

    def read_from_file(self, filename):
        if not os.path.exists(filename):
            raise FileNotFoundError(f'Error! The file {filename} was not found.')

        with open(filename, mode='r') as file:
            student_results = csv.DictReader(file)
        
            for row in student_results:
                row['score'] = int(row['score'])
                self.student_data.append(row)
                
    def __str__(self):
        return f'{self.student_data}'
    
    def __repr__(self):
        pass

def main():
    filename = 'student_results.csv'
    perfomance = PerfomanceAnalyser()
    perfomance.read_from_file(filename)
    print(perfomance)

if __name__ == '__main__':
    main()
```

# CEATE A CSV FILE

```python
# ---------- CREATE A CSV FILE ----------
import csv

students = [
    {"name": "Tlhompho", "surname": "Ratlhogo", "score": 37},
    {"name": "Moses", "surname": "Mabogwane", "score": 72},
    {"name": "Mmakgomo", "surname": "Seete", "score": 90},
    {"name": "Silas", "surname": "Rakwena", "score": 88},
    {"name": "Kagiso", "surname": "Makgata", "score": 65},
    {"name": "Khothatso", "surname": "Seeme", "score": 45}
]

filename = "student_results.csv"

with open(filename, mode='w', newline='') as file:
    fieldnames = ['name', 'surname', 'score']
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(students)

print(f"File '{filename}' has been created successfully!")
```

```python

students = [
    {'name': 'Alice', 'score': 20}, 
    {'name': 'Bob', 'score': 25},
    {'name': 'Charlie', 'score': 20}
]

def find_by_score(data_list, target_score):
    """
    Searches a list of dictionaries for a specific score 
    and prints the associated name and score.
    """
    found = False
    
    # Iterate through each dictionary in the list
    for entry in data_list:
        # Check if the 'score' key matches our target
        if entry['score'] == target_score:
            print(f"Match Found - Name: {entry['name']}, Score: {entry['score']}")
            found = True
            
    if not found:
        print(f"No entries found with a score of {target_score}.")

# 2. Implementation: Call the function
search_score = 20
find_by_score(students, search_score)
```
